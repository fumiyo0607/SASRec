```python
import gzip
from collections import defaultdict
from datetime import datetime
```

```python
def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)
```

```python
countU = defaultdict(lambda: 0)
countP = defaultdict(lambda: 0)
line = 0
```

```python
dataset_name = 'Beauty_5'
f = open('reviews_' + dataset_name + '.txt', 'w')
```

```python
for l in parse('reviews_' + dataset_name + '.json.gz'):
    line += 1
    f.write(" ".join([l['reviewerID'], l['asin'], str(l['overall']), str(l['unixReviewTime'])]) + ' \n')
    asin = l['asin']
    rev = l['reviewerID']
    time = l['unixReviewTime']
    countU[rev] += 1
    countP[asin] += 1
f.close()
```

```python
usermap = dict()
usernum = 0
itemmap = dict()
itemnum = 0
User = dict()
```

```python
for l in parse('reviews_' + dataset_name + '.json.gz'):
    line += 1
    asin = l['asin']
    rev = l['reviewerID']
    time = l['unixReviewTime']
    # 　評価件数が５件以下のものは無視    
    if countU[rev] < 5 or countP[asin] < 5:
        continue

    if rev in usermap:
        userid = usermap[rev]
    else:
        usernum += 1
        userid = usernum
        usermap[rev] = userid
        User[userid] = []
    if asin in itemmap:
        itemid = itemmap[asin]
    else:
        itemnum += 1
        itemid = itemnum
        itemmap[asin] = itemid
    User[userid].append([time, itemid])
```

```python
# sort reviews in User according to time
for userid in User.keys():
    User[userid].sort(key=lambda x: x[0])
```
